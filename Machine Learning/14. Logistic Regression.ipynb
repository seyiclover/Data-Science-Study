#### Logistic Function

```python
import numpy as np

z = np.arange(-10, 10, 0.01)
g = 1 / (1 + np.exp(-z))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(z, g)
plt.grid()
plt.show();
```

```python
plt.figure(figsize=(12,8))
ax = plt.gca()  # gca: 설정값을 변경할 수 있게 해 줌

ax.plot(z, g)
# 축 지정
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('center')
# 축 제거 
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

##### Logistic Regression Cost Function

```python
h = np.arange(0.01, 1, 0.01)

C0 = -np.log(1-h)
C1 = -np.log(h)

plt.figure(figsize=(12,8))
plt.plot(h, C0, label='y=0')
plt.plot(h, C1, label='y=1')
plt.legend()

plt.show()
```

#### 실습예제 

```python
import pandas as pd 

wine_url = "https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/wine.csv"

wine =  pd.read_csv(wine_url, index_col=0)
wine.head()
```

```python
# 맛 등급 만들어 넣기 
wine['taste'] = [1. if grade>5 else 0. for grade in wine['quality']]

X = wine.drop(['taste', 'quality'], axis=1)
y = wine['taste']
```

```python
# 데이터 분리 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.2, random_state=13)
```

```python
# 로지스틱 회귀 테스트 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression(solver='liblinear', random_state=13) # solver: 최적화 알고리즘 선택 // linear: 변수가 적을 때 주로 사용 
lr.fit(X_train, y_train)

y_pred_tr = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

```python
# 스케일러 적용해서 파이프라인 구축 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

estimators = [('scalar', StandardScaler()), 
              ('clf', LogisticRegression(solver='liblinear', random_state=13))]

pipe = Pipeline(estimators)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_tr = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

- 상승효과가 약간 있다. 

```python
# Decision Tree와 비교 
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
models = {'logistic regression':pipe, 'decision tree':wine_tree}
```

```python
for model_name, model in models.items():
    print(model_name)
    print(model)
```

```python
# AUC 그래프를 이용한 모델 간 비교 
from sklearn.metrics import roc_curve

plt.figure(figsize=(10,8))
plt.plot([0,1], [0,1])
for model_name, model in models.items():
    pred = model.predict_proba(X_test)[:, 1] # 모든 행의 두 번째 컬럼 
    fpr, tpr, thresholds = roc_curve(y_test, pred)
    plt.plot(fpr, tpr, label=model_name)

plt.grid()
plt.legend()
plt.show()
```

#### PIMA 인디언 당뇨병 예측

```python
import pandas as pd

PIMA_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/diabetes.csv'

PIMA = pd.read_csv(PIMA_url)
PIMA.head()
```

```python
PIMA.info()
```

```python
# float으로 데이터 변환 
PIMA = PIMA.astype('float')
PIMA.info()
```

```python
# 상관관계 확인 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(PIMA.corr(), cmap='YlGnBu')
plt.show()
```

##### 결측치 확인 방법

```python
# 컬럼 별로 0이 몇 개 있는지 확인 
(PIMA==0).astype(int).sum()
```

- 결측치는 데이터에 따라 그 정의가 다르다. 지금은 0이라는 숫자가 혈압에 있다는 것은 문제가 있다. 

- 의학적 지식과 PIMA 인디언에 대한 정보가 없으므로 평균값으로 대체

```python
zero_features = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']
PIMA[zero_features] = PIMA[zero_features].replace(0, PIMA[zero_features].mean())
(PIMA==0).astype(int).sum()
```

```python
# 데이터 나누기 
from sklearn.model_selection import train_test_split

X = PIMA.drop(['Outcome'], axis=1)
y = PIMA['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13, stratify=y)
```

```python
# Pipeline 만들기 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

estimators = [('scalar', StandardScaler()), 
              ('clf', LogisticRegression(solver='liblinear', random_state=13))]

pipe_lr = Pipeline(estimators)
pipe_lr.fit(X_train, y_train)
pred = pipe_lr.predict(X_test)
```

```python
from sklearn.metrics import (accuracy_score, recall_score, precision_score, roc_auc_score, f1_score)

print('Accuracy : ', accuracy_score(y_test, pred))
print('Recall : ', recall_score(y_test, pred))
print('Precision : ', precision_score(y_test, pred))
print('AUC score : ', roc_auc_score(y_test, pred))
print('f1 score : ', f1_score(y_test, pred))
```

```python
# 다변수 방정식의 각 계수 값 확인 
coeff = list(pipe_lr['clf'].coef_[0])
labels = list(X_train.columns)
```
