```python
import pandas as pd
```

```python
red_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-red.csv'
white_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-white.csv'

red_wine = pd.read_csv(red_url, sep=';')
white_wine = pd.read_csv(white_url, sep=';')
```

```python
red_wine.head()
```

```python
white_wine.head()
```

```python
white_wine.columns
```

```python
red_wine['color'] = 1.
white_wine['color'] = 0.

wine = pd.concat([red_wine, white_wine])
wine.info()
```

```python
wine.head()
```

```python
wine['quality'].unique()
```

```python
wine['quality'].value_counts()
```

```python
import plotly.express as px

fig = px.histogram(wine, x='quality')
fig.show()
```

```python
# 레드와 화이트 와인별 등급 
fig = px.histogram(wine, x='quality', color='color')
fig.show()
```

```python
X = wine.drop(['color'], axis=1)
y = wine['color']
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

np.unique(y_train, return_counts=True)
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Histogram(x=X_train['quality'], name='Train'))
fig.add_trace(go.Histogram(x=X_test['quality'], name='Test'))

fig.update_layout(barmode='overlay') # 겹침 설정 
fig.update_traces(opacity=0.75)
fig.show()
```

```python
# 결정나무 훈련 
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
# accuracy 확인 
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

```python
# 데이터 전처리 - MinMaxScalar와 StandardScaler
# boxplot 그리기 
fig = go.Figure()
fig.add_trace(go.Box(y=X['fixed acidity'], name='fixed acidity'))
fig.add_traces(go.Box(y=X['chlorides'], name='chlorides'))
fig.add_trace(go.Box(y=X['quality'], name='quality'))

fig.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

MMS = MinMaxScaler()
SS = StandardScaler()

SS.fit(X)
MMS.fit(X)

X_ss = SS.transform(X)  # numpy형태 
X_mms = MMS.transform(X)

X_ss_pd = pd.DataFrame(X_ss, columns=X.columns)  # 데이터 프레임 형태 
X_mms_pd = pd.DataFrame(X_mms, columns=X.columns)
```

```python
# MinMaxScaler
fig = go.Figure()
fig.add_traces(go.Box(y=X_mms_pd['fixed acidity'], name='fixed acidity'))
fig.add_traces(go.Box(y=X_mms_pd['chlorides'], name='chlorides'))
fig.add_traces(go.Box(y=X_mms_pd['quality'], name='quality'))

fig.show()
```

```python
# StandardScaler

def px_box(target_df):
    fig = go.Figure()
    fig.add_traces(go.Box(y=target_df['fixed acidity'], name='fixed acidity'))
    fig.add_traces(go.Box(y=target_df['chlorides'], name='chlorides'))
    fig.add_traces(go.Box(y=target_df['quality'], name='quality'))

    fig.show()
```

```python
px_box(X_mms_pd)
```

```python
# MinMaxScaler를 적용해서 다시 학습 

X_train, X_test, y_train, y_test = train_test_split(X_mms_pd, y, test_size=0.2, random_state=13)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```

```python
# StandardScaler를 적용해서 다시 학습 

X_train, X_test, y_train, y_test = train_test_split(X_ss_pd, y, test_size=0.2, random_state=13)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc : ', accuracy_score(y_train, y_pred_tr))
print('Test Acc : ', accuracy_score(y_test, y_pred_test))
```
