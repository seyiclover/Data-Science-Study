```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
```

```python
iris.data[:, 2:]  # 데이터 형태가 이중 리스트로 되어있기 때문에 :, 를 입력해주면 리스트 내부의 리스트에 접근   # petal leng와 width만 가져온다.
```

```python
iris_tree.fit(iris.data[:, 2:], iris.target) # fit은 학습 명령어 # 슬라이싱한 데이터에 iris.target이라는 정답을 알려주고 학습 진행
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
y_pred_tr
```

```python
iris.target
```

```python
accuracy_score(iris.target, y_pred_tr)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 6))
plot_tree(iris_tree);
```
