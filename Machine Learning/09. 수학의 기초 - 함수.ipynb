#### 다항함수

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 2, 100)  # -3부터 2까지 100개를 만든다. 
y = 3 * x**2 + 2

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('$x$')  # $를 넣어주면 이탤릭체로 표현된다. 
plt.ylabel('$3x^2 +2$')
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.style 

mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)
plt.show()
```

```python
# 다항함수의 x축 이동 
x = np.linspace(-5, 5, 100)
y1 = 3*x**2 + 2
y2 = 3*(x+1)**2 +2  # x 방향으로 -1만큼 이동 
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y1, lw=2, ls='dashed', label='$y=3x^2 +2$')  #lw: line width/ ls: line style
plt.plot(x, y2, label='$y=3(x+1)^2 +2$')
plt.legend(fontsize=15)
plt.xlabel('$x$', fontsize=25)
plt.xlabel('$y$', fontsize=25)
plt.show()
```

#### 지수함수

```python
x = np.linspace(-2, 2, 100)
a11, a12, a13 = 2, 3, 4
y11, y12, y13 = a11**x, a12**x, a13**x

a21, a22, a23 = 1/2, 1/3, 1/4
y21, y22, y23 = a21**x, a22**x, a23**x
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y11, color='k', label='$2^x$')
ax[0].plot(x, y12, '--', color='k', label='$3^x$')
ax[0].plot(x, y13, ':', color='k', label='$4^x$')

ax[0].legend(fontsize=20)

ax[1].plot(x, y21, color='k', label='$(1/2)^x$')
ax[1].plot(x, y22, '--', color='k', label='$(1/3)^x$')
ax[1].plot(x, y23, ':', color='k', label='$(1/4)^x$')

ax[1].legend(fontsize=20)

plt.show()
```

#### 로그함수

```python
def log(x, base):
    return np.log(x)/np.log(base)

x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.001, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)  # log10, log e(자연로그) 
y21, y22 = log(x1, 1/10), log(x2, 1/np.e) # 마이너스 승 
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x1, y11, color='k', label='$\log_{10} x$')
ax[0].plot(x2, y12, '--', color='k', label='$\log_{e} x$')

ax[0].set_xlabel('$x$', fontsize=25)
ax[0].set_ylabel('$y$', fontsize=25)
ax[0].legend(fontsize=20, loc='lower right')

ax[1].plot(x1, y21, color='k', label='$\log_{1/10} x$')
ax[1].plot(x2, y22, '--', color='k', label='$\log_{1/e} x$')

ax[1].set_xlabel('$x$', fontsize=25)
ax[1].set_ylabel('$y$', fontsize=25)
ax[1].legend(fontsize=20, loc='upper right')

plt.show()
```

#### 시그모이드

```python
z = np.linspace(-10, 10, 100)
sigma = 1/(1+np.exp(-z))

plt.figure(figsize=(12, 8))
plt.plot(z, sigma)
plt.xlabel('$z$', fontsize=25)
plt.ylabel('$\sigma(z)$', fontsize=25)
plt.show()
```

#### 벡터함수

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)
U, V = np.meshgrid(u, v)
X = U
Y = V
Z = (1 + U**2) + (V / (1 + V**2))
```

```python
# meshgrid
t = np.linspace(0, 4, 3)
p = np.linspace(0, 4, 3)
T, P = np.meshgrid(t, p)
T, P
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')  # 3차원 그래프 projection 

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

ax.scatter3D(X, Y, Z, marker='.', color='gray')
plt.show()
```

#### 합성함수

```python
x = np.linspace(-4, 4, 100)
y = x**3 - 15 * x + 30  # f(x)
z = np.log(y)  # g(y)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label='$x^3 - 15x + 30$', color='k')
ax[0].legend(fontsize=18)

ax[1].plot(y, z, label='$\log(y)$', color='k')
ax[1].legend(fontsize=18)

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, z, label='$\log(f(x))$', color='k')
ax[0].legend(fontsize=18)

ax[1].plot(x, y, label='$x^3 - 15x + 30$', color='k')
ax[1].legend(fontsize=18)
ax_tmp = ax[1].twinx()  # 2번째 그래프에서 x축을 하나 더 생성 
ax_tmp.plot(x, z, label='$\log(f(x))$', color='k')

plt.show()
```
